```python
import sys

sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('/Users/codyyu/Desktop/CMPE 297/chromedriver')
```

```python
driver.get('https://directory.nationalrestaurantshow.com/8_0/explore/exhibitor-gallery.cfm?featured=false')
```

```python
exhibitors = driver.find_elements("xpath", '//ul[@class="cards exh-featured"]')
```

```python
from selenium.webdriver.common.by import By
exhibitors = driver.find_elements(By.TAG_NAME, 'li')
```

```python
len(exhibitors)
```

```python
exhibitor_list = []
for exhibitor in exhibitors:

    if exhibitor.text != '':
        cur_exhibitor = []
        
        cur_exhibitor_name = ''
        cur_exhibitor_booth = ''
        cur_exhibitor_description = ''

        lines = exhibitor.text.split('\n')

        cur_exhibitor_name = lines[0]
        
        if (len(lines) > 1):
            cur_exhibitor_booth = lines[1]

        if (len(lines) > 2):
            cur_exhibitor_description = lines[2]
            
        cur_exhibitor.append(cur_exhibitor_name)
        cur_exhibitor.append(cur_exhibitor_booth)
        cur_exhibitor.append(cur_exhibitor_description)


        exhibitor_list.append(cur_exhibitor)
```

```python
keywords = ['package', 'box', 'packaging', 'box', 'bio', 'bowl', 'plate', 'fiber', 'green', 'tray', 'container', 'cup', 'plastic', 'foam']
```

```python
keyword_hit_list = []

for exhibitor in exhibitors:
    for keyword in keywords:
        if keyword in exhibitor.text.lower():
            cur_exhibitor = []
        
            cur_exhibitor_name = ''
            cur_exhibitor_booth = ''
            cur_exhibitor_description = ''

            lines = exhibitor.text.split('\n')

            cur_exhibitor_name = lines[0]
            
            if (len(lines) > 1):
                cur_exhibitor_booth = lines[1]

            if (len(lines) > 2):
                cur_exhibitor_description = lines[2]
                
            cur_exhibitor.append(cur_exhibitor_name)
            cur_exhibitor.append(cur_exhibitor_booth)
            cur_exhibitor.append(cur_exhibitor_description)


            keyword_hit_list.append(cur_exhibitor)
            break
```

```python
df = pd.DataFrame(keyword_hit_list) 
df = df.rename(columns={0: 'Name', 1: 'Booth', 2: 'Description'})
df
```
